La etapa BRONZE representa el primer nivel del proceso de transformación de datos dentro de un pipeline tipo Data Lakehouse.
Su objetivo principal es recibir, explorar y validar los datos en su forma más cruda (RAW), tal como llegan desde la fuente original.

En esta etapa no se realizan modificaciones profundas ni transformaciones avanzadas. En su lugar, nos enfocamos en:

✔ Comprender la estructura del dataset

Incluyendo tipos de datos, columnas disponibles y formato general.

✔ Evaluar la calidad del contenido

Como detectar valores nulos, inconsistencias, duplicados, rangos indebidos y posibles anomalías.

✔ Realizar un perfilamiento inicial

Generando estadísticas esenciales: conteo total de registros, mínimos, máximos, rangos, distribuciones por categoría, merchants, año, mes, hora, etc.

✔ Preparar la información para la siguiente capa (SILVER)

Ejecutando tareas ligeras como clasificaciones por rango, exploración de fechas o derivación de columnas simples, pero sin alterar la naturaleza original del dataset.

En resumen, la etapa BRONZE nos permite entender profundamente los datos antes de limpiarlos o transformarlos, asegurando que cada decisión en SILVER y GOLD esté basada en un análisis correcto del origen.
Es un paso crucial para garantizar calidad, trazabilidad y confiabilidad en todo el proceso analítico y de detección de fraude.

**MÓDULO 6 — Contar valores nulos por columna**

```python
query = """
SELECT
    COUNT(*) AS total_filas,
    COUNT(CC_NUM) AS cc_num_no_nulos,
    COUNT(MERCHANT) AS merchant_no_nulos,
    COUNT(CATEGORY) AS category_no_nulos,
    COUNT(AMT) AS amt_no_nulos,
    COUNT(FIRST_NAME) AS first_no_nulos,
    COUNT(LAST_NAME) AS last_no_nulos,
    COUNT(GENDER) AS gender_no_nulos,
    COUNT(STREET) AS street_no_nulos,
    COUNT(CITY) AS city_no_nulos,
    COUNT(STATE) AS state_no_nulos
FROM CARD_TRANSACTION.BRONZE.RAW_TRANSACTIONS
"""

df_nulls = pd.read_sql(query, connection)
df_nulls
```

**MÓDULO 7 — Rango de montos (CASE en SQL)**

```python
query = """
SELECT
    CASE
        WHEN AMT < 10 THEN '1. 0 - 10'
        WHEN AMT < 50 THEN '2. 10 - 50'
        WHEN AMT < 100 THEN '3. 50 - 100'
        WHEN AMT < 200 THEN '4. 100 - 200'
        WHEN AMT < 500 THEN '5. 200 - 500'
        WHEN AMT < 1000 THEN '6. 500 - 1000'
        WHEN AMT < 5000 THEN '7. 1000 - 5000'
        ELSE '8. 5000+'
    END AS rango_monto,
    COUNT(*) AS total_transacciones
FROM CARD_TRANSACTION.BRONZE.RAW_TRANSACTIONS
GROUP BY rango_monto
ORDER BY rango_monto;
"""

df_rangos = pd.read_sql(query, connection)
df_rangos
```

**MÓDULO 8 — Distribución por categoría**

```python
query = """
SELECT CATEGORY, COUNT(*) AS total_transacciones
FROM CARD_TRANSACTION.BRONZE.RAW_TRANSACTIONS
GROUP BY CATEGORY
ORDER BY total_transacciones DESC
"""

df_categoria = pd.read_sql(query, connection)
df_categoria
```

**MÓDULO 9 — Distribución por merchant**

```python
query = """
SELECT MERCHANT, COUNT(*) AS total_transacciones
FROM CARD_TRANSACTION.BRONZE.RAW_TRANSACTIONS
GROUP BY MERCHANT
ORDER BY total_transacciones DESC
LIMIT 50
"""

df_merchant = pd.read_sql(query, connection)
df_merchant
```

**MÓDULO 10 — Años presentes en la data**

```python
query = """
SELECT YEAR(TRANS_DATE_TRANS_TIME) AS anio,
       COUNT(*) AS total_transacciones
FROM CARD_TRANSACTION.BRONZE.RAW_TRANSACTIONS
GROUP BY anio
ORDER BY anio
"""

df_anios = pd.read_sql(query, connection)
df_anios
```

**MÓDULO 11 — Meses por año**

```python
query = """
SELECT
    YEAR(TRANS_DATE_TRANS_TIME) AS anio,
    MONTH(TRANS_DATE_TRANS_TIME) AS mes,
    COUNT(*) AS total_transacciones
FROM CARD_TRANSACTION.BRONZE.RAW_TRANSACTIONS
GROUP BY anio, mes
ORDER BY anio, mes
"""

df_mes = pd.read_sql(query, connection)
df_mes
```

**MÓDULO 12 — Distribución por hora del día**

```python
query = """
SELECT
    HOUR(TRANS_DATE_TRANS_TIME) AS hora,
    COUNT(*) AS total_transacciones
FROM CARD_TRANSACTION.BRONZE.RAW_TRANSACTIONS
GROUP BY hora
ORDER BY hora
"""

df_hora = pd.read_sql(query, connection)
df_hora
```

**MÓDULO 13 — MIN y MAX del monto**

7.-Valores nulos

En este módulo evaluamos la calidad del dataset identificando la cantidad de valores nulos por columna.
Dado que Snowpark no permite un .isna() estilo Pandas, realizamos el análisis columna por columna utilizando filter(col(c).is_null()).count().
Finalmente creamos un DataFrame Snowpark con los resultados.
Este paso es clave para identificar campos incompletos que requieren limpieza en la capa SILVER.

```python
# MÓDULO — Conteo de valores nulos por columna

columnas = df.columns

null_results = []

for c in columnas:
    nulos = df.filter(col(c).is_null()).count()
    null_results.append((c, nulos))

# Convertimos la lista en un DataFrame Snowpark
df_nulos = session.create_dataframe(null_results, schema=["COLUMNA", "NULOS"])

print("=== Conteo de valores nulos por columna ===")
df_nulos.show()
```

8.-Aquí calculamos estadísticas descriptivas del monto (AMT) usando Snowpark:

min() → valor mínimo

max() → valor máximo

avg() → promedio

Este análisis permite conocer la distribución general de los montos y detectar posibles valores atípicos o rangos relevantes del comportamiento transaccional.

```python
from snowflake.snowpark.functions import min, max, avg, col

df_stats = df.select(
    min(col("AMT")).alias("MIN_AMT"),
    max(col("AMT")).alias("MAX_AMT"),
    avg(col("AMT")).alias("AVG_AMT")
)

print("Estadísticas AMT:")
df_stats.show()
```

9.-En este módulo generamos columnas temporales derivadas de TRANS_DATE_TRANS_TIME usando funciones Snowpark como year(), month() y hour().
Esto es fundamental para el análisis temporal: detectar picos de actividad, meses más movidos o patrones horarios asociados a fraude.

```python
df_fecha = df.select(
    col("TRANS_DATE_TRANS_TIME"),
    year(col("TRANS_DATE_TRANS_TIME")).alias("ANIO"),
    month(col("TRANS_DATE_TRANS_TIME")).alias("MES"),
    hour(col("TRANS_DATE_TRANS_TIME")).alias("HORA")
)

df_fecha.show(5)
```

10.-Aquí construimos una columna categórica llamada RANGO_MONTO utilizando la función case() para agrupar montos en rangos definidos (0–10, 10–50, 50–100, etc.).
Posteriormente se realiza un conteo por rango.
Este paso es clave para la segmentación y para comprender cómo se distribuyen las transacciones por niveles de valor.

```python
from snowflake.snowpark.functions import col, year, month, hour, when, lit

df_rangos = df.select(
    when(col("AMT") < 10, lit("1. 0 - 10"))
    .when(col("AMT") < 50, lit("2. 10 - 50"))
    .when(col("AMT") < 100, lit("3. 50 - 100"))
    .when(col("AMT") < 200, lit("4. 100 - 200"))
    .when(col("AMT") < 500, lit("5. 200 - 500"))
    .when(col("AMT") < 1000, lit("6. 500 - 1000"))
    .when(col("AMT") < 5000, lit("7. 1000 - 5000"))
    .otherwise(lit("8. 5000+"))
    .alias("RANGO_MONTO")
).group_by("RANGO_MONTO").count().sort(col("COUNT").desc())

df_rangos.show()
```

11.-En este módulo agrupamos las transacciones por la columna CATEGORY y contamos cuántos registros hay por cada una.
Esto permite identificar qué categorías de compra aparecen con mayor frecuencia en los datos BRONZE (como “personal_care”, “home”, “kids_pets”, etc.).

```python
df_categoria = df.group_by(col("CATEGORY")).count().sort(col("COUNT").desc())

print("Distribución por categoría:")
df_categoria.show()
```

12.-Aquí se realiza un group_by("MERCHANT") seguido de count() y orden descendente para obtener los 20 comerciantes con mayor número de transacciones.
Este tipo de ranking es útil para análisis de negocio, detección de patrones o detección de comercios irregulares.

```python
df_merchants = df.group_by("MERCHANT")\
                  .count()\
                  .sort(col("COUNT").desc())\
                  .limit(20)

print(" TOP 20 MERCHANTS")
df_merchants.show()
```
